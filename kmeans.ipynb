```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs 
#used to crate isotopic gaussian blobs 
#isotopic -> = property of distributation having same characteristic or properties in all direction 


#gaussian blobs  = clusture or group of data points generated according to a gaussian distribution
# gaussian distribution = continous Probability distribution that is symetric and bell-shaped 
#each clusture is like blob 
#normal distributation = gaussian distribution 
```

```python
#parameters 
n_sample = 300 
n_feature = 2 
centres = 4 

#generate data 
import random 
random.random_state = 42
X , y = make_blobs(n_samples= n_sample  , n_features = n_feature  , centers = centres , cluster_std= 3   , random_state = 42 )

#visualize the data 
plt.scatter(X[: , 0 ] , X[: ,1]  , c = y  , cmap = 'viridis' , edgecolor = 'k' )
```

```python
random.random_state = 42

from sklearn.cluster import KMeans 
wcss = []
for i in range(1 , 11):
    kmean = KMeans(n_init = 10 , n_clusters = i )
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

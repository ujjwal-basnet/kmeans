```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.datasets import make_blobs 
#used to crate isotopic gaussian blobs 
#isotopic -> = property of distributation having same characteristic or properties in all direction 


#gaussian blobs  = clusture or group of data points generated according to a gaussian distribution
# gaussian distribution = continous Probability distribution that is symetric and bell-shaped 
#each clusture is like blob 
#normal distributation = gaussian distribution 
```

```python
#parameters 
n_sample = 300 
n_feature = 2 
centres = 4 

#generate data 
import random 
random.random_state = 42
X , y = make_blobs(n_samples= n_sample  , n_features = n_feature  , centers = centres , cluster_std= 3   , random_state = 42 )

#visualize the data 
plt.scatter(X[: , 0 ] , X[: ,1]  , c = y  , cmap = 'viridis' , edgecolor = 'k' )
```

```python
random.random_state = 42

from sklearn.cluster import KMeans 
wcss = []
for i in range(1 , 11):
    kmean = KMeans(n_init = 10 , n_clusters = i )
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
from sklearn.cluster import KMeans

wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_init = 10, n_clusters = i)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python
kmeans_ = KMeans(n_init = 10 , n_clusters= 4)
```

```python
y_kmeans = kmeans_.fit_predict(X)
```

```python
plt.scatter(X[y_kmeans  == 0 , 0 ]   , X[y_kmeans == 0 , 1] , s = 60 , c = 'red' , label = 'cluster 1')
plt.scatter(X[y_kmeans == 1 , 0 ] , X[y_kmeans == 1 , 1 ] , s = 60 , c='blue' , label = 'Cluster2')
plt.scatter(X[y_kmeans == 2 , 0] , X[y_kmeans == 2 , 1 ] , s = 60 , c = 'green' , label = 'cluster 3 ')
plt.scatter(X[y_kmeans == 3 , 0 ] , X[y_kmeans ==3 , 1] , s = 60 , c ='yellow' , label = 'cluster 4 ')
```
